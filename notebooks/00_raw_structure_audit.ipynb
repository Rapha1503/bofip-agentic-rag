# Raw Structure Audit

Notebook de lecture visuelle du clean-room BOFIP-only.

Objectifs:
- voir à quoi ressemblent les documents bruts retenus
- comparer XML, HTML et structure extraite
- comprendre les content types supportés ou problématiques
- vérifier visuellement titres, sections, paragraphes, tables, liens et références

```python
from pathlib import Path
import json
import sys
from pprint import pprint

PROJECT_ROOT = Path.cwd().resolve().parent
SRC_ROOT = PROJECT_ROOT / 'src'
if str(SRC_ROOT) not in sys.path:
    sys.path.insert(0, str(SRC_ROOT))

from bofip_cleanroom.jsonio import read_jsonl
from bofip_cleanroom.models import raw_document_from_dict

RAW_DOCS_PATH = PROJECT_ROOT / 'data' / 'interim' / 'raw_docs_sample_10.jsonl'
SUMMARY_PATH = PROJECT_ROOT / 'data' / 'reports' / 'phase1_extract_summary_sample_10.json'

docs = [raw_document_from_dict(item) for item in read_jsonl(RAW_DOCS_PATH)]
summary = json.loads(SUMMARY_PATH.read_text(encoding='utf-8'))
len(docs)
```
