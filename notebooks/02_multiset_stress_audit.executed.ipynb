# Multi-Set Stress Audit

Comparer plusieurs sets différents pour éviter de raisonner toujours sur le même échantillon.

```python
from pathlib import Path
import json

PROJECT_ROOT = Path.cwd().resolve().parent
REPORT_PATH = PROJECT_ROOT / 'data' / 'reports' / 'phase0_multiset_stress_audit.json'
report = json.loads(REPORT_PATH.read_text(encoding='utf-8'))
report['generated_at']
```

```python
[{
    'name': item['name'],
    'sample_size': item['sample_size'],
    'content_type_counts': item['content_type_counts'],
    'section_count_stats': item['section_count_stats'],
    'paragraph_count_stats': item['paragraph_count_stats'],
    'docs_with_zero_paragraphs': item['docs_with_zero_paragraphs'],
    'docs_with_synthetic_root_only': item['docs_with_synthetic_root_only'][:5],
} for item in report['sets']]
```

```python
for item in report['sets']:
    print('\n===', item['name'], '===')
    print('content types:', item['content_type_counts'])
    print('section stats:', item['section_count_stats'])
    print('paragraph stats:', item['paragraph_count_stats'])
    print('zero paragraphs:', item['docs_with_zero_paragraphs'])
    print('synthetic root only sample:', item['docs_with_synthetic_root_only'][:10])
```
